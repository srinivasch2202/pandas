```python
arr=np.random.randint(0,100,(5,3))

row=['mon','tue','wed','thrus','frid']
columnames=['jan','feb','march']
df=pd.DataFrame(arr,index=row,columns=columnames)

df
```

```python
my_dict={
    'data1':[1,2,3,4,5],
    'data2':[6,7,8,9,10],
    'data3':[6,7,8,9,10],
}
df=pd.DataFrame(my_dict)
df
# we can pass column and rows names to this dict by default it comes as mentioned 
```

```python
# but if we need different columns first we can mention 
col=['data3','data2','data1']
df=pd.DataFrame(my_dict,columns=col)
df
```
